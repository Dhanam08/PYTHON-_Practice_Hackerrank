```python
import pandas as pd
import matplotlib.pyplot as plt
result=pd.read_csv('HospInfo.csv')
result.groupby('State').count()[['Provider ID']]
result.plot(kind='bar',x='State',y='Provider ID')
```

```python
result.groupby(['Hospital Type']).count()[['Provider ID']]
result.plot(kind='bar',x='Hospital Type',y='Provider ID')
plt.show()
```

```python
result.groupby('Hospital Ownership').count()[['Hospital Name']]
result.plot(kind='bar',x='Hospital Ownership',y='Provider ID')
plt.show()
```
